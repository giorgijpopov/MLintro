# Шаг 1 - Загрузка данных

см https://docs.scipy.org/doc/numpy/reference/generated/numpy.loadtxt.html

```python
import numpy as np
X200 = np.loadtxt ('ecg_data/ecg200_samples.csv')
X208 = np.loadtxt ('ecg_data/ecg208_samples.csv')
X231 = np.loadtxt ('ecg_data/ecg231_samples.csv')
```

### Шаг 1.1(Опционально) - Подготовка данных (нормализация)

array / max(array), array - matrix row

# Шаг 2 - Загрузка разметки

```python
y200_char = np.loadtxt ('ecg_data/ecg200_labels.csv', delimiter='/n', dtype='str')
y231_char = np.loadtxt ('ecg_data/ecg231_labels.csv', delimiter='/n', dtype='str')

d = {'N' : 0, 'V' : 1, 'R' : 2}
def convert(x):
    return d[x]

vectorizer = np.vectorize(convert)
y200 = vectorizer(y200_char)
y231 = vectorizer(y231_char)
```

# Шаг 3 - Кластеризация комплексов

см: 
* http://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html
* http://scikit-learn.org/stable/modules/generated/sklearn.cluster.AgglomerativeClustering.html
* http://scikit-learn.org/stable/modules/generated/sklearn.cluster.DBSCAN.html

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Normalizer

from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.cluster import AgglomerativeClustering

def search_clusters (X, alg, scaler):
    scaler.fit (X)
    X_scaled = scaler.transform (X)
    return alg.fit_predict(X_scaled)

clusters231 = search_clusters (X231, AgglomerativeClustering (n_clusters=2), MinMaxScaler ())
```

```python
def create_obvious_emissions (X):
    for i in range (X.shape[0]):
        if np.abs (X[i]).max () > 0.9:
            X[i] = np.full ((1, X.shape[1]), 100)
            
X200_copy = X200.copy ()
create_obvious_emissions (X200_copy)

clusters200 = search_clusters (X200_copy, AgglomerativeClustering (n_clusters=3), StandardScaler ())
```

```python
clusters208_kmeans = search_clusters (X208, KMeans (n_clusters=2), MinMaxScaler ())
clusters208_agglom = search_clusters (X208, AgglomerativeClustering (n_clusters=3), MinMaxScaler ())
```

# Шаг 4 - Визуализация полученного разбиения:

```python
import matplotlib.pyplot as plt
from sklearn.metrics.cluster import v_measure_score
from sklearn.metrics.cluster import silhouette_score

                
def draw_clusters (X, y, axes, row, titles):
    samples_count = X.shape[0]
    fetch_count = X.shape[1]
    unique = np.unique (y)
    max_graphs = 50
    graphs_count = np.zeros (unique.size)
    x_points = np.linspace(0, 100, fetch_count)

    for cluster in range (unique.size):
        ax = axes[row, cluster]
        for i in range (samples_count):
            if y[i] == unique[cluster] and graphs_count[cluster] < max_graphs:
                ax.plot (x_points, X[i, :], color='red')
                ax.set_title (str (titles[cluster]))
                graphs_count[cluster] += 1
                
def draw_info (X, y, clusters, y_titles, cluster_titles):
    y_unique = np.unique (y)
    clusters_unique = np.unique (clusters)
    plot_rows_count = 2
    plot_columns_count = max (y_unique.size, clusters_unique.size)
    fig, axes = plt.subplots(nrows=plot_rows_count, ncols=plot_columns_count, figsize=(12, 7))
    draw_clusters (X, y, axes, 0, y_titles)
    draw_clusters (X, clusters, axes, 1, cluster_titles)

draw_info (X200, y200, clusters200, np.unique (y200_char), np.unique (clusters200))

plt.tight_layout()
```

```python
draw_info (X231, y231, clusters231, np.unique (y231_char), np.unique (clusters231))

plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=2, ncols = max (np.unique (clusters208_kmeans).size, 
                                               np.unique (clusters208_agglom).size), figsize=(12, 7))
draw_clusters (X208, clusters208_kmeans, axes, 0, np.unique (clusters208_kmeans))
draw_clusters (X208, clusters208_agglom, axes, 1, np.unique (clusters208_agglom))

plt.tight_layout()
```
